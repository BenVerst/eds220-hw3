```python
### ===== FINAL CODE =====

# Imported libraries
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# loaded in spills and county data
spills = pd.read_csv('https://data.ny.gov/api/views/u44d-k5fk/rows.csv?date=20231106&accessType=DOWNLOAD&sorting=true')
counties = gpd.read_file('data/tl_2022_us_county.shp')

# Changed column names to be lower case
spills.columns = spills.columns.str.lower()

# Changed column names so spaces are replaced by _
spills.columns = spills.columns.str.replace(" ", "_")

# Changed spill_date column to be a datetime datatype
spills.spill_date = pd.to_datetime(spills.spill_date)

# Filtered dates and material type
petr_23 = spills.loc[(spills['spill_date'] >= '2023-1-1') & # Starts January 1 2023
                     (spills['spill_date'] <= '2023-10-31') & # Ends October 31 2023
                     (spills['material_family'] == 'Petroleum')] # Filters for Petroleum in material_family

# Grouped county and checked observations counts
spill2 = petr_23.groupby(['county']).count()

# Filtered spill_number column
spills_per_county = spill2[['spill_number']]

# Renamed column to n_spills and reset index
spills_per_county = spills_per_county.rename(columns = {'spill_number' : 'n_spills'}).reset_index()

# Changed column names to be lower case
counties.columns = counties.columns.str.lower()

# Filtered for New York
ny_counties = counties[counties.statefp == '36']

# Selected the first item in tuple using [0]
st_lawrence_index = spills_per_county[spills_per_county.county=='St Lawrence'].index[0]

# Renamed St Lawrence to St. Lawrence
spills_per_county.at[st_lawrence_index, 'county'] = 'St. Lawrence'

# Rename ny_counties column 'name' to 'county_name'
ny_counties = ny_counties.rename(columns={'name':'county_name'})

# Renamed spills_per_county column 'county' to 'county name'
spills_per_county = spills_per_county.rename(columns = {'county':'county_name'})

# Merged both datasets by overwriting ny_counties
ny_counties = pd.merge(ny_counties, 
                            spills_per_county,
                            how='inner', # inner join
                            on='county_name') # merged by 'county_name'

# Added 1 extra value due to Tuscarora Indian Reservation not being accounted for
ny_counties.at[45, 'n_spills'] += 1


# Created empty figure
fig, ax = plt.subplots(figsize=(10, 6))

ax.axis('off') # turn off axis
ax.set_title('Number of Petroleum Spills Per County in New York (2023)', # add title
fontsize=17) # font size

# Added New York county to map
ny_counties.plot(ax=ax,
               color='none', # no color
               edgecolor = 'black') # border color

# Added n_spills to map
ny_counties.plot(ax=ax, 
                  column='n_spills', # n_spills data column
                  legend=True, # adds legend
                  cmap='Greens', # hue color
                  alpha=0.7, # transparency
                  legend_kwds={'label': "Number of Spills"}) # legend label

# Annotated data
ax.annotate("Data: New York State Department of Environmental Conservation, accessed Nov 7, 2023 \nhttps://data.ny.gov/Energy-Environment/Spill-Incidents/u44d-k5fk", 
            xy=(0.01, .07), # text position
            xycoords='figure fraction', # coordinates have a range of 0-1
            fontsize=9) # font size

# Additional annotation indicating one spill occurred in Tuscarora Indian Reservation within Niagara county
ax.annotate("One additional spill occurred in Tuscarora Indian Reservation within Niagara county", 
            xy=(0.01, .03), # text position
            xycoords='figure fraction', # coordinates have a range of 0-1
            fontsize=9) # font size
```
